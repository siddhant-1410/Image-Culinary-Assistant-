```python
import requests
from bs4 import BeautifulSoup
from reportlab.lib.pagesizes import letter
from reportlab.pdfgen import canvas
from textwrap import wrap  # For wrapping long text

# Headers to mimic a browser request
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/119.0.0.0 Safari/537.36"
}

# Function to scrape a recipe
def scrape_recipe(dish_name):
    base_url = "x"
    dish_url = base_url + dish_name.replace(" ", "-") + "/"
    
    response = requests.get(dish_url, headers=headers)

    if response.status_code == 200:
        soup = BeautifulSoup(response.text, "html.parser")

        # Extract recipe title
        title = soup.find("h1", class_="entry-title")
        recipe_title = title.get_text(strip=True) if title else dish_name

        # Extract ingredients
        ingredients = [ingredient.get_text(strip=True) for ingredient in soup.select("ul.wprm-recipe-ingredients li")]

        # Extract instructions
        instructions = [step.get_text(strip=True) for step in soup.select("div.wprm-recipe-instruction-text")]

        # Return extracted data
        return {
            "dish": recipe_title,
            "url": dish_url,
            "ingredients": ingredients,
            "instructions": instructions
        }

    else:
        print(f"âŒ Failed to fetch recipe for {dish_name} (Status: {response.status_code})")
        return None

# List of dishes to scrape
dishes = [
    "paneer-butter-masala", "dal-makhani", "biryani-rice", "matar-paneer", "samosa",
    "rasgulla-recipe-how-to-make-rasgullas", "chole-bhature", "tomato-dal-andhra-tomato-dal",
    "khaman-dhokla-microwave-recipe", "dosa-recipe-dosa-batter-recipe", "jalebi-recipe-how-to-make-jalebi",
    "veg-kathi-roll-recipe", "pani-puri-recipe-mumbai-pani-puri-recipe", "instant-pot-pav-bhaji",
    "medu-vada-recipe-medu-vada", "rotis-made-from-whole-wheat-flour", "gulab-jamun-recipe-with-khoya",
    "gajar-halwa-recipe-gajar-ka-halwa", "idli-recipe-how-to-make-soft-idlis", "veg-noodles-recipe-easy-veg-noodles",
    "poori-a-kind-of-fried-indian-bread", "aloo-gobi-recipe-punjabialoo-gobi",
    "aloo-tikki-aloo-pattice-recipe-made-from-leftover-potatoes", "bhindi-masala",
    "dal-makhani-restaurant-style-recipe", "kadai-paneer-recipe-kadai-paneer",
    "litti-chokha-recipe", "palak-paneer"
]

# Scrape recipes
recipes = []
for dish in dishes:
    print(f"ğŸ” Scraping: {dish.replace('-', ' ').title()}")
    recipe_data = scrape_recipe(dish)
    if recipe_data:
        recipes.append(recipe_data)

# Function to save recipes as a PDF with proper text wrapping
def save_to_pdf(recipes, filename="recipes.pdf"):
    c = canvas.Canvas(filename, pagesize=letter)
    y = 750  # Initial Y position

    def draw_wrapped_text(text, x, y, max_width, line_height):
        """Helper function to wrap long text and print it properly."""
        lines = wrap(text, width=max_width)  # Wrap text based on width
        for line in lines:
            c.drawString(x, y, line)
            y -= line_height
            if y < 50:  # If text reaches the bottom, go to new page
                c.showPage()
                y = 750
        return y

    for recipe in recipes:
        c.setFont("Helvetica-Bold", 14)
        c.drawString(100, y, recipe["dish"])
        y -= 20
        
        c.setFont("Helvetica", 12)
        c.drawString(100, y, "Ingredients:")
        y -= 20
        for ingredient in recipe["ingredients"]:
            y = draw_wrapped_text(f"- {ingredient}", 120, y, 80, 15)
        
        c.drawString(100, y, "Instructions:")
        y -= 20
        for idx, step in enumerate(recipe["instructions"], start=1):
            y = draw_wrapped_text(f"{idx}. {step}", 120, y, 80, 15)
        
        y -= 30  # Space between recipes
        
        # If Y position is too low, create a new page
        if y < 50:
            c.showPage()
            y = 750
    
    c.save()
    print("âœ… Recipes saved to recipes.pdf")

# Save recipes to PDF
save_to_pdf(recipes)
```
